```
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")
data.head()
```

```
data.shape
```

```
#Plotting graph between x and y
data.plot(kind='scatter', x='Hours', y='Scores')
plt.show()
```

```
 data.columns
```

```
#Splitting the data into training and testing set
train,test = train_test_split(data, test_size = 0.25, random_state = 123)
```

```
train_x = train.drop("Scores",axis=1)
train_y = train["Scores"]
```

```
test_x = test.drop("Scores",axis=1)
test_y = test["Scores"]
```

```
 # Modelling the data
 lr = LinearRegression()
 lr.fit(train_x, train_y)
```

```
lr.coef_
```

```
lr.intercept_
```

```
# Plotting the regression line
line = lr.coef_*train_x+lr.intercept_

# Plotting for the test data
plt.scatter(train_x, train_y)
plt.plot(train_x, line);
plt.show()
```

```
# Predicting the test set 
pre = lr.predict(test_x)
```

```
# Comparing actual v/s predicted scores
compare_df = pd.DataFrame({'Actual Score': test_y, 'Predicted score':pre})
print(compare_df)
```

```
# Evaluating the model using mean squared error
mean_squared_error(test_y,pre,squared=False)
```

```
#Predicting the target by giving new value
hour = [9.25]
new_pr=lr.predict([hour])
print("Number of Hours = {}".format([hour]))
print("Predicted Score = {}".format(new_pr[0]))
```
